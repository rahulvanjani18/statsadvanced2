1. What is hypothesis testing in statistics?
> Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data, by testing an assumed claim (null hypothesis) against an alternative hypothesis.
---
2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
> The **null hypothesis (H₀) states there is no effect or no difference.
The **alternative hypothesis (H₁) states there is an effect or a difference.
---
3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
>The significance level is the probability of rejecting the null hypothesis when it is true.
It sets the cutoff for deciding whether test results are statistically significant.
---
4. What are Type I and Type II errors? Give examples of each.
> Type I error  Rejecting a true null hypothesis. Example – Concluding a drug works when it actually doesn’t.

  Type II error  Failing to reject a false null hypothesis. Example – Concluding a drug doesn’t work when it actually does.

---

5.What is the difference between a Z-test and a T-test? Explain when to use
each.
> Z-test: Used when population variance is known and sample size is large (n     >30).
   T-test: Used when population variance is unknown and sample size is small (n ≤  30).

```python
'''
7.Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
'''


import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50
alpha = 0.05

x_bar = np.mean(sample_data)
sigma = np.std(sample_data, ddof=0)
n = len(sample_data)

z = (x_bar - mu_0) / (sigma / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", x_bar)
print("Z-statistic:", z)
print("p-value:", p_value)


if p_value < alpha:
    print("Reject Null Hypothesis: Sample mean is significantly different from 50.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from 50.")
'''
Interpretation:
If `p-value < 0.05`, the sample mean is significantly different from 50.
If `p-value ≥ 0.05`, there is not enough evidence to conclude a difference.
'''
```

```python
'''
8.Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


data = np.random.normal(loc=50, scale=5, size=100)


mean = np.mean(data)
std_err = stats.sem(data)


confidence = 0.95
ci = stats.t.interval(confidence, df=len(data)-1, loc=mean, scale=std_err)

print("Sample Mean:", mean)
print("95% Confidence Interval:", ci)


plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label=f"Lower CI = {ci[0]:.2f}")
plt.axvline(ci[1], color='blue', linestyle='dashed', linewidth=2, label=f"Upper CI = {ci[1]:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```

```python
'''
9.Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
'''
import numpy as np
import matplotlib.pyplot as plt


def zscore_visualization(data):
    mean = np.mean(data)
    std = np.std(data, ddof=0)
    z_scores = (data - mean) / std

    print("Z-scores:", z_scores)

    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.legend()
    plt.show()


data = np.array([10, 12, 13, 15, 18, 20, 22, 25, 30])
zscore_visualization(data)
```
